```python
import lyricsgenius

api_key = "0U9DJ07Oxm77Re6NU4yZChDXy4vjSe1DNchnbeUIWz_bmb5WawXYQR3iPEEYg7VP"

def get_lyrics(artist_name, track_name, api_key)->str:
    genius = lyricsgenius.Genius(api_key, remove_section_headers=True)
    try:
        song = genius.search_song(title=track_name, artist=artist_name)
        if song:
            # print(f"Lyrics for '{track_name}' by {artist_name}:\n")
            # print(song.lyrics)
            return song.lyrics
        else:
            print("Lyrics not found. Please check the artist or track name.")
    except Exception as e:
        print(f"An error occurred: {e}")

```
